```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def extract_data(name, first_rows = 1):
    df = pd.read_csv(name)
    for i in range(first_rows):
        df = df.drop(i, axis=0)
    df = df.drop("Unnamed: 0", axis=1)
    display(df)

    x, y = [], []

    for _, j in df.iterrows():
        for index, value in j.items():
            x.append(int(index))
            y.append(value)
    return np.array(x), np.array(y)
```

```python
x, y = extract_data("experiments_results/experiment1/kernels.txt")
ax = plt.subplot()
sns.lineplot(x, y, ax = ax)
ax.set_xlabel("Размерность тензора", fontsize=13)
ax.set_ylabel("Время выполнения алгоритма", fontsize=13)
```

```python
x, y = extract_data("experiments_results/experiment1/modes.txt")
ax = plt.subplot()
sns.lineplot(x, y, ax = ax)
ax.set_xlabel("Размер моды тензора", fontsize=13)
ax.set_ylabel("Время выполнения алгоритма", fontsize=13)
```

```python
x, y = extract_data("experiments_results/experiment1/ranks.txt", 2)
ax = plt.subplot()
sns.lineplot(x, y / (x ** 4), ax = ax)
ax.set_xlabel("Начальный ранг (ранг округления, R)", fontsize=13)
ax.set_ylabel("Время исполнения программы / R^4", fontsize=13)
```
