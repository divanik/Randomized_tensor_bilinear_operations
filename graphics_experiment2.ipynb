```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def extract_data(name, first_rows = 1):
    df = pd.read_csv(name)
    for i in range(first_rows):
        df = df.drop(i, axis=0)
    df = df.drop("Unnamed: 0", axis=1)
    display(df)

    x, y = [], []

    for _, j in df.iterrows():
        for index, value in j.items():
            x.append(int(index))
            y.append(value)
    return np.array(x), np.array(y)
```

```python
x1, y1 = extract_data("experiments_results/experiment2/increasing_rounding_rank_approx.txt")
x2, y2 = extract_data("experiments_results/experiment2/increasing_rounding_rank_precise.txt")
```

```python
ax = plt.subplot()
sns.lineplot(x1, y1, ax = ax, label = "Optimized version")
sns.lineplot(x2, y2, ax = ax, label = "Naive version")
ax.set_xlabel("Ранг округления", fontsize=13)
ax.set_ylabel("Время выполнения алгоритма", fontsize=13)
```
